```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import gc
```

```python
admision= pd.read_csv('admission.csv', delimiter= ';')
admision= admision.drop(['Unnamed: 0'], axis=1)
```

```python
datasignal = pd.read_csv('signal.csv', delimiter= ';')
datasignal= datasignal.drop(['Unnamed: 0', 'Unnamed: 0.1'], axis=1)
```

```python
age=pd.read_csv('age.csv', delimiter= ';')
age= age.drop(['Unnamed: 0'], axis=1)
```

```python
df_median = pd.DataFrame(datasignal.groupby(['pat_id','day','parameter'])['value'].median())
df_median.columns = ['median']
df_median.reset_index(drop=False,inplace=True)
df_mean = pd.DataFrame(datasignal.groupby(['pat_id','day','parameter'])['value'].mean())
df_mean.columns = ['mean']
df_mean.reset_index(drop=False,inplace=True)
df_std = pd.DataFrame(datasignal.groupby(['pat_id','day','parameter'])['value'].std())
df_std.columns = ['std']
df_std.reset_index(drop=False,inplace=True)
```

```python

merged = pd.merge(df_median,df_mean, on=['pat_id','day','parameter'])
merged = pd.merge(merged,df_std, on=['pat_id','day','parameter'])

del [df_median, df_mean, df_std]
gc.collect()
```

```python
datasignal = pd.read_csv('signal.csv', delimiter= ';')
datasignal= datasignal.drop(['Unnamed: 0', 'Unnamed: 0.1'], axis=1)
```

```python
datasignal.pivot(index=None, columns='parameter', values='value')
```

```python
data=pd.merge(datasignal, datasignal.pivot(index=None, columns='parameter', values='value'), left_index= True, right_index= True)
```

```python
data= data.drop(['parameter', 'value'], axis=1)
```

```python
data= data.fillna(value=0, axis=None, inplace=False)
```

```python
df_median = pd.DataFrame(datasignal.groupby(['pat_id','day','parameter'])['value'].median())
df_median.columns = ['median']
df_median.reset_index(drop=False,inplace=True)
df_mean = pd.DataFrame(datasignal.groupby(['pat_id','day','parameter'])['value'].mean())
df_mean.columns = ['mean']
df_mean.reset_index(drop=False,inplace=True)
df_std = pd.DataFrame(datasignal.groupby(['pat_id','day','parameter'])['value'].std())
df_std.columns = ['std']
df_std.reset_index(drop=False,inplace=True)
```

```python
df_data = pd.DataFrame(data.groupby(['pat_id','day'])['blood_pressure','respiration_rate','temperature'].mean())
df_data.columns = ['blood_pressure','respiration_rate','temperature']
df_data.reset_index(drop=False,inplace=True)
```
